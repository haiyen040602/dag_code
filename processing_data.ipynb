```python
def extract_number(text):
    match = re.search(r"\d+\.?\d*", text)
    if match:
        return int(match.group())
    else:
        return None
```

```python
duration_groupby = meta_package.groupby(by='duration')['package_name'].apply(list).to_dict()
# duration_groupby = list(duration_groupby.values)
duration_list = dict()
for item in duration_groupby:
    if extract_number(item) is not None:
        duration_list[extract_number(item)] = duration_groupby[item]

print(duration_list)
```

```python
keys = []
for key, values in duration_list.items():
    keys.append(key)

print(len(keys))
print(keys)
```

```python
string = "-	7GB/ngày abc xyz"
pattern = r"(?<!\w)(\d+\.?\d?)(\s*)GB(\s*)/(\s*)(ngày|tháng|chu kỳ)(?!\w)"
# pattern  = r"(?<!\w)(\d+\.?\d?)(\s*)GB(\s*)/(\s*)(ngày|tháng|chu kỳ)(?!\w)"
# pattern  = r"(\d+\sGB/\w+)" #r"(?<!\d)\d+ (?:GB|GB/ngày|GB/tháng)(?!\d)" #r"(\d+\sGB/\w+)"
# matches = re.findall(pattern, string)

def find_regex(string, pattern):
    matches = re.findall(pattern, str(string))
    return matches

print()
```

```python
data_usage1 = pd.read_csv('../database/metadata/data_usage1.csv')
data_usage2 = pd.read_csv('../database/metadata/data_usage2.csv')
data_usage2 = data_usage2.dropna(how='all')
data_usage1 = data_usage1.dropna(how='all')

data_usage = pd.concat([data_usage1, data_usage2])
data_usage.describe
```

```python
data_usage.dtypes
```

```python
data_usage['data'] = data_usage['content'].apply(lambda x: find_regex(x, pattern=pattern))
data_usage = data_usage.dropna(subset=['duration'])
data_usage
```

```python
code_list = []
for index, row in data_usage.iterrows():
    
    if len(row['data']):
        data = row['data'][0]
        dur = extract_number(row['duration'])
        if (data[-1] == 'ngày' and dur == 1) or data[-1] == 'chu kỳ':
            usage = float(data[0]) * 10**9
        elif data[-1] == 'ngày' and dur is not None:
            usage = float(data[0]) * dur * 10**9
        else:
            usage = 0

    tup = (row['code'], usage)
    code_list.append(tup)
            
print(code_list)

    # print(data)
    # data = 
    # tup = (row['code'], )
```

## Count duration used per month

```python
from ast import literal_eval
import numpy as np

def match_duration(string_list, series):
    ret_list = []
    if not pd.isnull(string_list):
        string_list = literal_eval(string_list)
        for string in string_list:
            for label, value in series.items():
                if isinstance(value, list) and string in value:
                    ret_list.append(label)
    return ret_list
```

```python
service_cols =  ['register_service_n_0', 'register_service_n_1', 'register_service_n_2']

service_data = raw_data
print(type(service_data))

for col in service_cols:
    # service_data[col] = raw_data[col].to_list()
    service_data[col] = service_data[col].apply(lambda x : match_duration(x, duration_list))
service_data.head()
```

```python
durations = [30, 60, 90, 180, 270, 360, 420, 540, 720]
import pandas as pd
import collections
def count_values_in_cells(raw, list_cols, durations):
  df = raw.copy()
  for index in raw.index:
    for col in list_cols:
      all_values = collections.Counter(df[col][index])
      # print(all_values)
      for val, count in all_values.items():
          # print(val, count)
          if val in durations:
            new_col_name = f'{col}_{val}'
            if new_col_name not in df.columns:
              df[new_col_name] = 0
            df.at[index, new_col_name] = count

  #   df.drop(col, axis=1, inplace=True)
  return df
```

```python
updated_df = count_values_in_cells(service_data, service_cols, durations)
print(updated_df.columns)
```

```python
for ser in service_cols:
    for d in durations:
        col_name  = f"{ser}_{d}"
        if col_name not in updated_df.columns:
            updated_df[col_name] = 0

print(updated_df.columns)
```

```python
def get_duration(col_name):
    if(col_name == ''):
        return 0 
    list_col =  col_name.split(',')
    durs = []
    for col in list_col:
        splited = col.split('_')
        durs.append(int(splited[-1]))
    
    return int(max(durs)/30)

get_duration('')
```
